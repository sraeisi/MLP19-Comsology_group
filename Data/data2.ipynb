```python
import h5py
import numpy as np
import pandas as pd
f = h5py.File('groups_135.Subhalo.SubhaloPos.hdf5' , 'r') #data includes galaxy position in simulation box for 135th snapshot
head_subhalo = f['Subhalo']
data_subhalo = head_subhalo['SubhaloPos']
X_pos_subhalo = data_subhalo[:,0]
Y_pos_subhalo = data_subhalo[:,1]
Z_pos_subhalo = data_subhalo[:,2]
a = pd.DataFrame(data_subhalo[:,:] , columns = 'X_coordinate Y_coordinate Z_coordinate'.split()) #converting hdf5 to pandas dataframe
a.head(10)

```

```python
print('we do not have missing data and the data is scaled. we have one feature that  is position in 3d space.')
```

```python
#plotting  2d histogram of galaxy position
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.hist2d(X_pos_subhalo, Y_pos_subhalo ,norm = mpl.colors.LogNorm(), bins = 64)
plt.xlim([0,75000])
plt.ylim([0,75000])
plt.title('histogram of galaxy Coordinates')
plt.xlabel('X_position[ckpc/h]')
plt.ylabel('Y_position[ckpc/h]')
plt.colorbar()
plt.show()
```

```python
#plotting 3d distribution of galaxies in space
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111, projection = '3d')
X_label = 'X_position'
Y_label = 'Y_position'
z_label = 'Z_position'
plt_title = 'Galaxy positions in simulation box'
ax.scatter(X_pos_subhalo , Y_pos_subhalo , Z_pos_subhalo , zdir = 'z' , s = 0.5,  c = 'blue'  )

plt.show()
```

```python
import h5py    #data includes dark matter position in simulation box. goal of the project is mapping between darkmatter position and galaxy position 
import numpy as np
#we have 8 hdf5 files. we upload it and then merge it
f = h5py.File('new1.h5' , 'r')
head = f['DM']
data = head['Coordinates0']
X_pos_0 = data[:,0]
Y_pos_0 = data[:,1]
Z_pos_0 = data[:,2]

data = head['Coordinates1']
X_pos_1 = data[:,0]
Y_pos_1 = data[:,1]
Z_pos_1 = data[:,2]

data = head['Coordinates2']
X_pos_2 = data[:,0]
Y_pos_2 = data[:,1]
Z_pos_2 = data[:,2]

data = head['Coordinates3']
X_pos_3 = data[:,0]
Y_pos_3 = data[:,1]
Z_pos_3 = data[:,2]

data = head['Coordinates4']
X_pos_4 = data[:,0]
Y_pos_4 = data[:,1]
Z_pos_4 = data[:,2]

data = head['Coordinates5']
X_pos_5 = data[:,0]
Y_pos_5 = data[:,1]
Z_pos_5 = data[:,2]

data = head['Coordinates6']
X_pos_6 = data[:,0]
Y_pos_6 = data[:,1]
Z_pos_6 = data[:,2]

data = head['Coordinates7']
X_pos_7 = data[:,0]
Y_pos_7 = data[:,1]
Z_pos_7 = data[:,2]


```

```python
#merging hdf5 files
X_pos = np.hstack((X_pos_0,X_pos_1,X_pos_2,X_pos_3,X_pos_4,X_pos_5,X_pos_6,X_pos_7))
Y_pos = np.hstack((Y_pos_0,Y_pos_1,Y_pos_2,Y_pos_3,Y_pos_4,Y_pos_5,Y_pos_6,Y_pos_7))
Z_pos = np.hstack((Z_pos_0,Z_pos_1,Z_pos_2,Z_pos_3,Z_pos_4,Z_pos_5,Z_pos_6,Z_pos_7))

```

```python
#ploting 2d histogram of dark matter position
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.hist2d(X_pos , Y_pos , norm = mpl.colors.LogNorm() , bins = 64)
plt.title('histogram of Dark matter particle position')
plt.xlabel('X_position[ckpc/h]')
plt.ylabel('Y_position[ckpc/h]')
plt.colorbar()
plt.show()
```
